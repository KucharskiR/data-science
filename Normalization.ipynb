```python
import numpy as np
import pandas as pd
np.set_printoptions(precision=12, suppress=True, linewidth=150)
pd.options.display.float_format = '{:.6f}'.format
# sns.set()   <--- seaborn
print(np.__version__)
```

```python
close = 75
vwamp = 70
```

```python
data = np.array([75,70,74,73,50,49.7,70,75,80,80])
```

```python
data = np.reshape(data,(-1,2))
data
```

```python
# df = pd.DataFrame(my_array, columns=['Column_A', 'Column_B', 'Column_C'])
df = pd.DataFrame(data, columns=['Close', 'VWAP'])
df
```

```python
for i in range(len(df)):
  multi = 10
  close = df.iloc[i,0]
  vwap = df.iloc[i,1]
  if(vwap<close):
    vwap /= close
    vwap-= 1
    vwap *= multi
  else:
    vwap = close/vwap
    vwap-= 1
    vwap *= (-1)*multi

  print(vwap)
```

### Heiken

```python
h = np.array([75,75.1,75.1,75,79,79.3,79.3,79])
h = h.reshape(-1,2)
h
```

```python
hdf = pd.DataFrame(h, columns=['H1', 'H2'])
hdf
```

```python
for i in range(len(hdf)):
  h1 = hdf.iloc[i,0]
  h2 = hdf.iloc[i,1]
  if(h1 < h2):
    heikenResult = (h1 / h2) -1
    heikenResult *= 100
  else:
    heikenResult = (h2 / h1) -1
    heikenResult *= -100

  print(heikenResult)
```
